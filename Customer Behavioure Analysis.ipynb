customer behaviour analysis

1. Project Overview
This project analyzes customer shopping behavior using transactional data from 3,900
purchases across various product categories. The goal is to uncover insights into spending
patterns, customer segments, product preferences, and subscription behavior to guide strategic
business decisions.


2. Dataset Summary
- Rows: 3,900
- Columns: 18
- Key Features:
- Customer demographics (Age, Gender, Location, Subscription Status)
- Purchase details (Item Purchased, Category, Purchase Amount, Season, Size, Color)
- Shopping behavior (Discount Applied, Promo Code Used, Previous Purchases, Frequency of
Purchases, Review Rating, Shipping Type)
- Missing Data: 37 values in Review Rating column

Data Preporcessing

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python

# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

 Exploratory Data Analysis 

```python
import plotly.express as px
```

1. Revenue Contribution by Gender
Business Question: Which gender contributes more revenue

```python
gender_revenue = df.groupby("gender", as_index=False)["purchase_amount"].sum()

fig = px.bar(
    gender_revenue,
    x="gender",
    y="purchase_amount",
    title="Total Revenue by Gender",
    text_auto=True
)
fig.show()
```

2. Revenue by Product Category

```python
category_revenue = (
    df.groupby("category", as_index=False)["purchase_amount"]
    .sum()
    .sort_values(by="purchase_amount", ascending=False)
)

fig = px.bar(
    category_revenue,
    x="category",
    y="purchase_amount",
    title="Revenue by Product Category",
    text_auto=True
)
fig.show()
```

3. Average Purchase Amount by Age

```python
age_spend = df.groupby("age", as_index=False)["purchase_amount"].mean()

fig = px.line(
    age_spend,
    x="age",
    y="purchase_amount",
    title="Average Purchase Amount by Age",
    markers=True
)
fig.show()
```

4. Seasonal Revenue Trend

```python
season_revenue = df.groupby("season", as_index=False)["purchase_amount"].sum()

fig = px.bar(
    season_revenue,
    x="season",
    y="purchase_amount",
    title="Season-wise Revenue",
    text_auto=True
)
fig.show()
```

5. Subscription vs Non-Subscription Revenue

```python
sub_revenue = df.groupby("subscription_status", as_index=False)["purchase_amount"].sum()

fig = px.pie(
    sub_revenue,
    names="subscription_status",
    values="purchase_amount",
    title="Revenue Split: Subscription vs Non-Subscription"
)
fig.show()
```

6. Discount Impact on Average Spending

```python
discount_avg = df.groupby("discount_applied", as_index=False)["purchase_amount"].mean()

fig = px.bar(
    discount_avg,
    x="discount_applied",
    y="purchase_amount",
    title="Average Purchase Amount: Discount Impact",
    text_auto=True
)
fig.show()
```

7.Preferred Payment Methods

8. Review Rating vs Purchase Amount

```python
fig = px.scatter(
    df,
    x="review_rating",
    y="purchase_amount",
    title="Review Rating vs Purchase Amount",
    hover_data=["category", "item_purchased"]
)
fig.show()
```

 9. Purchase Frequency vs Average Spend

```python
freq_spend = df.groupby("frequency_of_purchases", as_index=False)["purchase_amount"].mean()

fig = px.bar(
    freq_spend,
    x="frequency_of_purchases",
    y="purchase_amount",
    title="Average Spend by Purchase Frequency",
    text_auto=True
)
fig.show()
```

10. Revenue by Shipping Type

```python
shipping_revenue = df.groupby("shipping_type", as_index=False)["purchase_amount"].sum()

fig = px.bar(
    shipping_revenue,
    x="shipping_type",
    y="purchase_amount",
    title="Revenue by Shipping Type",
    text_auto=True
)
fig.show()
```
